```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Boston Housing Dataset 1978.xlsx")
```

```python
data.head()
```

```python
import statsmodels.api as sm
```

```python
X_rm = data["RM"].values.reshape(-1, 1)
X_lstat = data["LSTAT"].values.reshape(-1, 1)
```

```python
unos = np.ones((len(data), 1))
RM_LSTAT   = (data["RM"] * data["LSTAT"]).values.reshape(-1, 1)

X = np.column_stack((unos, X_rm, X_lstat, RM_LSTAT))
Y = data["MEDV"].values

ols = sm.OLS(Y, X)
results = ols.fit()
print(results.summary())
```

```python
df = data_z.copy()

for col in df.columns:
    if col == "MEDV":
        continue  

    x = df[col].to_numpy(dtype=float)

    x_min = np.min(x)
    x_max = np.max(x)
    aux = x - x_min

    rng = x_max - x_min
    if rng == 0:
        aux = np.zeros_like(x)      
    else:
        aux = aux / rng

    aux = np.clip(aux, 0.0, 1.0)

    df[col] = aux

df.head()
```

```python
X_rm = df["RM"].values.reshape(-1, 1)
X_lstat  = df["LSTAT"].values.reshape(-1, 1)
RM_LSTAT = (df["RM"] * df["LSTAT"]).values.reshape(-1, 1)
unos     = np.ones((len(df), 1))

X = np.column_stack((unos, X_rm, X_lstat, RM_LSTAT))
y = df["MEDV"].values

ols = sm.OLS(y, X)              
results = ols.fit()

print(results.summary())
```
